```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.preprocessing import LabelEncoder
from sklearn.naive_bayes import GaussianNB
from sklearn import metrics
```

```python
data=pd.read_csv("id3_ds.csv")
data.head()
```

```python
x=data[['weather','temp']]
y=data['play']
print(x)
```

```python
x1=x.apply(LabelEncoder().fit_transform)
print(x1)
```

```python
le=preprocessing.LabelEncoder()
y1=le.fit_transform(y)
print(y1)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x1,y1,test_size=0.33,random_state=42)
print('Training data shape: ',x_train.shape)
print('Testing data shape: ',x_test.shape)
```

```python
navie_model=GaussianNB()
navie_model.fit(x_train,y_train)
predictions=navie_model.predict(x_test)
print(metrics.classification_report(y_test,predictions))
data=pd.DataFrame(metrics.confusion_matrix(y_test,predictions),index=['yes','no'],columns=['yes','no'])
print(data)
```

```python
print(metrics.accuracy_score(y_test,predictions))
```
